```python
sns.FacetGrid(iris, hue = 'Species',size=5) \
    .map(plt.scatter, 'SepalLengthCm','SepalWidthCm').add_legend()
```

```python
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
```

```python
ax = sns.boxplot(data=iris, x = 'Species',y = 'PetalLengthCm')
ax = sns.stripplot(data=iris, x='Species', y='PetalLengthCm', jitter=True, edgecolor='green')
```

```python
sns.violinplot(x='Species', y = 'SepalLengthCm', data=iris)
```

```python
sns.violinplot(x='Species',y='PetalLengthCm', data=iris, size=6) 
#Denser regions of the data are fatter, and sparser thiner.
```

```python
sns.violinplot(x='Species',y='PetalWidthCm', data=iris, size=6) 
```

```python
sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
```

***OBSERVATION***

1.The dataset consists of 150 rows and 6 columns of data.

2.The column 'ID' is of no significance.

3.The columns/features:

                      Sepal-Width(cm): specifies the Sepal width in cemtimeters.
                      
                      Sepal-Length(cm): specifies the Sepal length in cemtimeters. 
                      
                      Petal-Length(cm): specifies the Petal length in cemtimeters.
                       
                      Petal-Width(cm): specifies the Petal width in cemtimeters.
                       
                       
4.The classes of the species of iris flower are 'Iris-setosa','Iris-versicolor','Iris-virginica'.

5.The given dataset is a balanced dataset i.e all the three species has 50 rows of data each.

6.From the above plots the category of species 'Iris-setosa' can be linearly seperated from 'Iris-versicolor' and 'Iris-virginica' using the feature  Petal-Length(cm).
